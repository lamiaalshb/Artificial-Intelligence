**2.Handling Outliers**

----  Handling Outliers Using Z-Score Analysis ----

```python
from scipy import stats


# ------------------ 1. Remove outliers using IQR ------------------
def remove_outliers_iqr(df):
    """
    Remove outliers using the Interquartile Range (IQR) method.
    """
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df >= lower_bound) & (df <= upper_bound)]

# ------------------ 2. Remove outliers using Z-score ------------------
def remove_outliers_zscore(df, threshold=3):
    """
    Remove outliers using the Z-score method.
    """
    z_scores = stats.zscore(df.select_dtypes(include=[np.number]))  # Calculate Z-scores
    return df[(np.abs(z_scores) < threshold).all(axis=1)]  # Remove rows with Z-scores > threshold

# ------------------ 3. Replace outliers with median ------------------
def replace_outliers_with_median(df):
    """
    Replace outliers with the median for each column.
    """
    for column in df.select_dtypes(include=[np.number]).columns:
        median = df[column].median()  # Calculate median
        std_dev = df[column].std()  # Calculate standard deviation
        df[column] = df[column].apply(lambda x: median if abs(x - df[column].mean()) > 3 * std_dev else x)
    return df

# ------------------ 4. Replace outliers with mean ------------------
def replace_outliers_with_mean(df):
    """
    Replace outliers with the mean for each column.
    """
    for column in df.select_dtypes(include=[np.number]).columns:
        mean = df[column].mean()  # Calculate mean
        std_dev = df[column].std()  # Calculate standard deviation
        df[column] = df[column].apply(lambda x: mean if abs(x - df[column].mean()) > 3 * std_dev else x)
    return df

# ------------------ 5. Clip outliers using Clipping ------------------
def clip_outliers(df):
    """
    Clip outliers using the Clipping method (capping values).
    """
    for column in df.select_dtypes(include=[np.number]).columns:
        lower_bound = df[column].quantile(0.05)
        upper_bound = df[column].quantile(0.95)
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# ------------------ Apply different outlier handling methods ------------------

# 1. Remove outliers using IQR
data_iqr = data.copy()
data_iqr = remove_outliers_iqr(data_iqr)

# 2. Remove outliers using Z-score
data_zscore = data.copy()
data_zscore = remove_outliers_zscore(data_zscore)

# 3. Replace outliers with median
data_median = data.copy()
data_median = replace_outliers_with_median(data_median)

# 4. Clip outliers using Clipping
data_clipped = data.copy()
data_clipped = clip_outliers(data_clipped)

# ------------------ Print statistics before and after processing ------------------

# Statistics before processing
print("### Statistics Before Processing ###")
print(data.describe())

# Statistics after removing outliers using IQR
print("\n### Statistics After IQR Processing ###")
print(data_iqr.describe())

# Statistics after removing outliers using Z-score
print("\n### Statistics After Z-score Processing ###")
print(data_zscore.describe())

# Statistics after replacing outliers with median
print("\n### Statistics After Median Replacement ###")
print(data_median.describe())

# Statistics after clipping outliers
print("\n### Statistics After Clipping ###")
print(data_clipped.describe())
```

The provided code defines several outlier removal techniques (IQR, Z-Score, Median/Mean Replacement, and Clipping) and applies them to a DataFrame, printing the summary statistics before and after each processing step.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import StandardScaler

# Log transformation of the price and volume data
data['close_SAR_log'] = np.log(data['close_SAR'] + 1)
data['close_USD_log'] = np.log(data['close_USD'] + 1)
data['volume_log'] = np.log(data['volume'] + 1)

# Distribution plotting before and after the log transformation
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.histplot(data['close_SAR'], kde=True, ax=axes[0], color='blue')
axes[0].set_title('Distribution of close_SAR (Original)', fontsize=14)

sns.histplot(data['close_SAR_log'], kde=True, ax=axes[1], color='green')
axes[1].set_title('Distribution of close_SAR (Log Transformed)', fontsize=14)

sns.histplot(data['volume_log'], kde=True, ax=axes[2], color='purple')
axes[2].set_title('Distribution of volume (Log Transformed)', fontsize=14)

plt.tight_layout()
plt.show()

# Outlier Removal Using IQR (Interquartile Range)
Q1 = data[['close_SAR_log', 'close_USD_log', 'volume_log']].quantile(0.25)
Q3 = data[['close_SAR_log', 'close_USD_log', 'volume_log']].quantile(0.75)
IQR = Q3 - Q1

# Define the boundaries (lower & upper)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
data_cleaned = data[~((data[['close_SAR_log', 'close_USD_log', 'volume_log']] < lower_bound) |
                       (data[['close_SAR_log', 'close_USD_log', 'volume_log']] > upper_bound)).any(axis=1)]

# Distribution plot after removing outliers
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.histplot(data_cleaned['close_SAR_log'], kde=True, ax=axes[0], color='blue')
axes[0].set_title('Distribution of close_SAR (Log Transformed & Cleaned)', fontsize=14)

sns.histplot(data_cleaned['close_USD_log'], kde=True, ax=axes[1], color='green')
axes[1].set_title('Distribution of close_USD (Log Transformed & Cleaned)', fontsize=14)

sns.histplot(data_cleaned['volume_log'], kde=True, ax=axes[2], color='purple')
axes[2].set_title('Distribution of volume (Log Transformed & Cleaned)', fontsize=14)

plt.tight_layout()
plt.show()


# 3. Data Scaling

scaler = StandardScaler()
data_scaled = data_cleaned.copy()

# Scaling the continuous columns
data_scaled[['close_SAR_log', 'close_USD_log', 'volume_log']] = scaler.fit_transform(data_cleaned[['close_SAR_log', 'close_USD_log', 'volume_log']])

# Plotting the scaled data
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.histplot(data_scaled['close_SAR_log'], kde=True, ax=axes[0], color='blue')
axes[0].set_title('Scaled close_SAR_log', fontsize=14)

sns.histplot(data_scaled['close_USD_log'], kde=True, ax=axes[1], color='green')
axes[1].set_title('Scaled close_USD_log', fontsize=14)

sns.histplot(data_scaled['volume_log'], kde=True, ax=axes[2], color='purple')
axes[2].set_title('Scaled volume_log', fontsize=14)

plt.tight_layout()
plt.show()

# Print statistics before and after processing
print("### Statistics Before Processing ###")
print(data[['close_SAR', 'close_USD', 'volume']].describe())

print("\n### Statistics After IQR Processing ###")
print(data_cleaned[['close_SAR_log', 'close_USD_log', 'volume_log']].describe())

print("\n### Statistics After Scaling ###")
print(data_scaled[['close_SAR_log', 'close_USD_log', 'volume_log']].describe())
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Assume your data is in a DataFrame 'df' containing the required columns
# 'close_SAR', 'close_USD', 'volume'
scaler = MinMaxScaler()

# Scale the data between 0 and 1
scaled_data = scaler.fit_transform(df[['close_SAR', 'close_USD', 'volume']])

# Function to create sequences of data for time series prediction
def create_sequences(data, time_step):
    X, y = [], []
    for i in range(len(data) - time_step):
        # Inputs: Sequence of values for all 3 columns
        X.append(data[i:i + time_step, :])  # All columns as inputs
        # Outputs: Future values for all 3 columns
        y.append(data[i + time_step, :])  # All columns as outputs
    return np.array(X), np.array(y)

# Define the time window (e.g., 60 days)
time_step = 60
X, y = create_sequences(scaled_data, time_step)

# Print shapes to verify
print(f"X shape: {X.shape}")  # Should be (number of sequences, time window, number of features)
print(f"y shape: {y.shape}")  # Should be (number of sequences, number of features)

# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Check shapes after splitting
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

Summary: The three-dimensional shape of the data is crucial for RNN models to effectively capture temporal dependencies and understand the sequence of historical data. For time series forecasting, the data needs to be structured in this format (samples, time steps, features) to allow the model to process the sequential nature of the data and make accurate predictions. This format ensures that the model can recognize patterns over time and use them for forecasting future values.



# LSTM (Long Short-Term Memory) MODEL

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt

# Build the model
model_LSTM = Sequential()  # Naming the model as model_LSTM
model_LSTM.add(LSTM(units=128, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2]), kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model_LSTM.add(Dropout(0.3))  # Dropout rate to reduce overfitting
model_LSTM.add(LSTM(units=64, return_sequences=False, kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model_LSTM.add(Dropout(0.3))  # Dropout rate to reduce overfitting
model_LSTM.add(Dense(64, activation='relu'))
model_LSTM.add(Dense(3, activation='linear'))  # The last layer of the regression

# Compile the model
model_LSTM.compile(optimizer=Adam(learning_rate=0.0005), loss='mean_squared_error', metrics=['mae'])

# Callbacks for learning rate adjustment and early stopping
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Model summary
model_LSTM.summary()

# Train the model
history = model_LSTM.fit(
    X_train, y_train,
    validation_data=(X_train, y_train),
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping, lr_scheduler],
    verbose=1
)

# Plotting the loss and MAE curves
plt.figure(figsize=(14, 5))

# Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()

# MAE
plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.title('MAE Curve')
plt.legend()

plt.show()
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import GRU, Dense

# Load the data
df = pd.read_csv('/content/dc.csv')

# Select the columns we will work with
df = df[['Unnamed: 0', 'close_SAR', 'close_USD', 'volume']]

# Convert the date to the correct type
df['Unnamed: 0'] = pd.to_datetime(df['Unnamed: 0'])
df.set_index('Unnamed: 0', inplace=True)

# Prepare the data for the model
# Split the data into inputs (X) and targets (y)
X = df[['close_SAR', 'close_USD', 'volume']].values
y = df[['close_SAR', 'close_USD', 'volume']].values  # We want to predict all three columns

# Scaling using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y)

# Split the data into training and test sets
train_size = int(len(X_scaled) * 0.8)
X_train, X_test = X_scaled[:train_size], X_scaled[train_size:]
y_train, y_test = y_scaled[:train_size], y_scaled[train_size:]

# Reshape the data to fit GRU input [samples, time steps, features]
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Build the GRU model
model = Sequential()
model.add(GRU(units=50, activation='linear', input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(3))  # Multiple outputs (SAR close price, USD close price, and volume)
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=1)

# Predict using the test set
predictions = model.predict(X_test)

# Inverse scaling for predictions
predictions_rescaled = scaler.inverse_transform(predictions)

# Inverse scaling for the true values
y_test_rescaled = scaler.inverse_transform(y_test)

# Plot predictions compared to true values
plt.figure(figsize=(10, 6))

# Plot the predictions and true values for SAR close price
plt.subplot(3, 1, 1)
plt.plot(predictions_rescaled[:, 0], label='Predicted Close Price (SAR)')
plt.plot(y_test_rescaled[:, 0], label='True Close Price (SAR)')
plt.legend()
plt.title('Predicted vs Actual Close Price (SAR)')

# Plot the predictions and true values for USD close price
plt.subplot(3, 1, 2)
plt.plot(predictions_rescaled[:, 1], label='Predicted Close Price (USD)')
plt.plot(y_test_rescaled[:, 1], label='True Close Price (USD)')
plt.legend()
plt.title('Predicted vs Actual Close Price (USD)')

# Plot the predictions and true values for volume
plt.subplot(3, 1, 3)
plt.plot(predictions_rescaled[:, 2], label='Predicted Volume')
plt.plot(y_test_rescaled[:, 2], label='True Volume')
plt.legend()
plt.title('Predicted vs Actual Volume')

plt.tight_layout()
plt.show()

# Calculate the difference between true values and predictions
comparison_df = pd.DataFrame({
    'True Close Price (SAR)': y_test_rescaled[:, 0],
    'Predicted Close Price (SAR)': predictions_rescaled[:, 0],
    'True Close Price (USD)': y_test_rescaled[:, 1],
    'Predicted Close Price (USD)': predictions_rescaled[:, 1],
    'True Volume': y_test_rescaled[:, 2],
    'Predicted Volume': predictions_rescaled[:, 2]
})

comparison_df['Difference (SAR)'] = comparison_df['True Close Price (SAR)'] - comparison_df['Predicted Close Price (SAR)']
comparison_df['Difference (USD)'] = comparison_df['True Close Price (USD)'] - comparison_df['Predicted Close Price (USD)']
comparison_df['Difference (Volume)'] = comparison_df['True Volume'] - comparison_df['Predicted Volume']

# Display the first 10 rows with the difference
print(comparison_df.head(10))

```

### Project Overview: Time Series Analysis for Bitcoin Price Prediction (USD & SAR) and Volume

This project focuses on time series forecasting for Bitcoin's closing price in both **USD** and **SAR**, as well as the trading **volume**. The goal is to analyze historical data to predict future Bitcoin prices and volumes, utilizing different machine learning models such as **Linear Regression**, **GRU**, and **LSTM**.

#### Key Points:
1. **Target Variables**:  
   - **Closing Price (USD)**: The price of Bitcoin in USD at the close of each trading day.
   - **Closing Price (SAR)**: The price of Bitcoin in SAR (Saudi Riyal), also at the close of each trading day.
   - **Volume**: The trading volume of Bitcoin for each day.

2. **Time Series Data**:  
   - The data used in this project consists of historical Bitcoin data for both **USD** and **SAR**, with attributes like **open**, **high**, **low**, **close**, and **volume** for each day.
   - The model aims to forecast the closing prices (both USD and SAR) and trading volume based on previous data.

3. **Model Approach**:  
   - We implemented different models to predict the future values of Bitcoin's **closing price** and **volume**:
     - **Linear Regression**: A simple yet effective model for predicting continuous data based on linear relationships.
     - **GRU (Gated Recurrent Units)**: A type of Recurrent Neural Network (RNN) that is well-suited for time series data due to its ability to capture temporal dependencies.
     - **LSTM (Long Short-Term Memory)**: Another type of RNN, more advanced than GRU, designed to remember long-term dependencies in sequential data.

4. **Evaluation**:  
   - The performance of the models was evaluated using metrics like **MAE (Mean Absolute Error)**, **RMSE (Root Mean Squared Error)**, and **R²** (Coefficient of Determination).
   - **GRU** showed the best performance in terms of accuracy and prediction error, followed by **LSTM**. **Linear Regression** performed decently but with higher prediction error.

5. **Objective**:  
   - The primary objective is to accurately predict the future **closing prices** (both USD and SAR) and **volume** of Bitcoin, providing useful insights for traders, investors, and analysts. This can help in making data-driven decisions regarding Bitcoin's market trends.

In conclusion, this project is a time series forecasting model applied to Bitcoin's historical data with the goal of predicting future trends in the cryptocurrency market. It highlights the comparison between traditional models like **Linear Regression** and more advanced models such as **GRU** and **LSTM**, with a focus on accuracy and performance in predicting **price** and **volume** trends.
